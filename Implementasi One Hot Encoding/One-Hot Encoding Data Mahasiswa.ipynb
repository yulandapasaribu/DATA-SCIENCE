Konversi ke numerik

```python
import pandas as pd

# Menghapus baris pertama jika mengandung NaN (hanya untuk kolom yang seluruhnya NaN)
data = data.dropna(how='all')

# Pilih hanya kolom yang isinya seluruhnya hanya "Ya" atau "Tidak"
binary_columns = [col for col in data.columns if set(data[col].dropna().unique()).issubset({"Ya", "Tidak"})]

# Konversi ke numerik: "Ya" -> 1, "Tidak" -> 0
data[binary_columns] = data[binary_columns].replace({"Ya": 1, "Tidak": 0})

# Tampilkan hanya kolom yang telah dikonversi, tetapi tanpa menampilkan baris yang memiliki NaN
print(data[binary_columns].dropna().head())  # Menggunakan dropna() hanya pada saat menampilkan
```

3. Cek Kategori Nilai Rata-rata

```python
columns_to_display = [
    'Berapa rata-rata peringkat Anda di kelas pada jenjang SD(Sekolah Dasar)?',
    'Berapa rata-rata peringkat Anda di kelas pada jenjang SMP( Sekolah Menengah Pertama )?',
    'Berapa rata-rata peringkat Anda di kelas pada jenjang SMA( Sekolah Menengah Atas  )?',
    'Berapa rata-rata nilai Matematika Anda pada Ujian Sekolah (nilai Ijazah) di SMA?',
    'Berapa rata-rata nilai Fisika Anda pada Ujian Sekolah (nilai Ijazah) di SMA? ( jika sebelumnya berasal dari IPS atau SMK bisa memilih range D)',
    'Berapa rata-rata nilai Biologi Anda pada Ujian Sekolah (nilai Ijazah)  di SMA?  ( jika sebelumnya berasal dari IPS atau SMK bisa memilih range D)',
    'Berapa rata-rata nilai Kimia Anda pada Ujian Sekolah (nilai Ijazah)  di SMA?  ( jika sebelumnya berasal dari IPS atau SMK bisa memilih range D)',
]

# Hilangkan spasi dalam nama kolom
data.columns = data.columns.str.strip()

# Menampilkan kolom yang dipilih
print(data[columns_to_display])
```

Konversi ke numerik dengan cara mengambil nilai rata-ratanya

```python
rank_mapping = {
    'Peringkat 1 - 3': 2,
    'Peringkat 4 - 10': 7,
    'Peringkat 11 - 20': 15
}

grade_mapping = {
    '93 - 100 (A)': 96.5,
    '83 - 92 (B)': 87.5,
    '75 - 83 (C)': 79
}

for col in columns_to_display:
    data[col] = data[col].map(rank_mapping).fillna(data[col].map(grade_mapping))
```

Konversi menjadi numerik dengan cara mengambil nilai rata-rata

```python
import pandas as pd
import numpy as np


data = pd.DataFrame({
    "Berapa lama waktu yang Anda habiskan menggunakan laptop setiap hari?": [
        np.nan, "9 - 11 jam", "6 - 8 jam", "6 - 8 jam", "3 - 5 jam",
        "12 - 13 jam", "3 - 5 jam", "6 - 8 jam", "12 - 13 jam", "6 - 8 jam"
    ],
    "Berapa lama waktu yang Anda habiskan menggunakan media sosial melalui Laptop setiap hari?": [
        np.nan, "3 - 5 jam", np.nan, "0 - 2 jam", np.nan,
        "0 - 2 jam", "3 - 5 jam", np.nan, "6 - 8 jam", np.nan
    ],
    "Berapa lama waktu yang Anda habiskan menggunakan media sosial melalui Ponsel setiap hari?": [
        np.nan, "6 - 8 jam", "3 - 5 jam", "3 - 5 jam", "3 - 5 jam",
        "12 - 13 jam", "3 - 5 jam", "0 - 2 jam", "6 - 8 jam", "0 - 2 jam"
    ]
})

# Dictionary untuk mapping rentang waktu ke nilai rata-rata
time_mapping = {
    "0 - 2 jam": 1.0,
    "3 - 5 jam": 4.0,
    "6 - 8 jam": 7.0,
    "9 - 11 jam": 10.0,
    "12 - 13 jam": 12.5
}

# Daftar kolom yang akan dikonversi
columns_to_convert = list(time_mapping.keys())

# Mengonversi rentang waktu ke nilai rata-rata
for col in data.columns:
    data[col] = data[col].map(time_mapping)

# Mengisi NaN dengan median masing-masing kolom
data.fillna(data.median(), inplace=True)

# Menampilkan hasil
print(data)
```
