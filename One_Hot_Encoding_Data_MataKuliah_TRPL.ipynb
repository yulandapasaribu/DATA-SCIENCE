```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Data numerik
numerical_data = {
    'Kode Mata Kuliah': [4143201, 4143202, 4143203, 4143204],
    'Nama Mata Kuliah': ["Algoritma", "Struktur Data", "Machine Learning", "PBO"],
    'Jumlah SKS': [3, 3, 3, 2],
    'Semester': [6, 6, 6, 6],
    'Jumlah Mahasiswa': [50, 60, 58, 70]
}

# Data kategorik
categorical_data = {
    "Dosen Pengampu": [
        "Oppir Hutapea, S.Tr.Kom, M.Kom",
        "Dr. Arnaldo Marulitua Sinaga, ST., M.InfoTech.",
        "Riyanthi Angrainy Sianturi, M.Ds.",
        "Oppir Hutapea, S.Tr.Kom, M.Kom"
    ],
    "Tipe Mata Kuliah": [
        "Wajib", "Pilihan", "Wajib", "Pilihan"
    ],
    "Ruang Kelas": [
        "GD516", "GD517", "GD513", "GD516"
    ],
    "Hari Perkuliahan": [
        "Selasa", "Senin", "Senin", "Selasa"
    ],
    "Waktu Perkuliahan": [
        "08.00 - 09.50", "10.00 - 11.50", "09.00 - 09.50", "08.00 - 09.50"
    ]
}

# Membuat DataFrame
df_numerical = pd.DataFrame(numerical_data)
df_categorical = pd.DataFrame(categorical_data)

# One-hot encoding untuk data kategorik
encoder = OneHotEncoder()
encoded_categorical = encoder.fit_transform(df_categorical).toarray()

# Membuat DataFrame dari hasil encoding dengan nama kolom yang lebih rapi
feature_names = encoder.get_feature_names_out(df_categorical.columns)

# Fungsi untuk merapikan nama kolom (menghapus prefiks kategori)
def clean_column_names(names):
    return [name.split('_')[1] for name in names]

cleaned_feature_names = clean_column_names(feature_names)
encoded_df = pd.DataFrame(encoded_categorical, columns=cleaned_feature_names)

# Menggabungkan data numerik dan hasil encoding kategorik
final_df = pd.concat([df_numerical, encoded_df], axis=1)

# Menampilkan hasil akhir
print(final_df)
```
